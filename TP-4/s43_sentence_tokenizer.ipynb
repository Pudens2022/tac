# Segmentation en phrases

## Imports

```python
import os
import sys
import nltk
from nltk.tokenize import sent_tokenize
```

## Fichiers d'inputs et d'outputs

```python
#infile = "../data/camille_Crise_economique/KB_JB838_1929-06-01_01-00001.txt"
outfile = "../data/sents.txt"
data_path = "../data/camille_Crise_economique/"
files = [f for f in sorted(os.listdir(data_path))]
```

## Segmentation en phrases du corpus complet et cr√©ation d'un nouveau fichier

**Important** : pour traiter le corpus complet, indiquez `LIMIT = None`

```python
LIMIT = 1000000
```

```python
with open(outfile, 'w', encoding="utf-8") as output:
    #content_list = []
    for myFile in files:
        with open(data_path+myFile, 'r', encoding='utf-8') as f:
            print(myFile) 
            content = f.readlines()
            content = content[:LIMIT] if LIMIT is not None else content
            n_lines = len(content)
            for i, line in enumerate(content):
                if i % 100 == 0:
                    print(f'processing line {i}/{n_lines}')
                sentences = sent_tokenize(line)
                for sent in sentences:
                    output.write(sent + "\n")
print("Done")
```
